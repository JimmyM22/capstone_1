```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv('heart.csv')
```

```python
#intro text
print("This data set contains 303 patients. For each patient, 14 attributes revolving the patient's current health. I will be focusing on five attributes. The last attributed indicates if the patient has heart disease. According to UCI's website, the orignial   dataset was donated over 30 years ago on 1988-07-01 and almost all the paitients are over 30.") 
print()
print("Hypothesis: Increase in age, cholestrol, and blood sugar levels as well being male would indicate a higher risk for heart      disease.")
print()

print("Sources:")
print("Data file taken from: https://www.kaggle.com/ronitf/heart-disease-uci/downloads/heart.csv/1")
print("The original source of the data: https://archive.ics.uci.edu/ml/datasets/heart+Disease")
print("Attribute clarifications by user IntiPic: https://www.kaggle.com/ronitf/heart-disease-uci/discussion/105877#latest-612520")

print()

print("Acknowledgements")
print("Creators:")
print("1. Hungarian Institute of Cardiology. Budapest: Andras Janosi, M.D.")
print("2. University Hospital, Zurich, Switzerland: William Steinbrunn, M.D.") 
print("3. University Hospital, Basel, Switzerland: Matthias Pfisterer, M.D.")
print("4. V.A. Medical Center, Long Beach and Cleveland Clinic Foundation: Robert Detrano, M.D., Ph.D.")

print("Donor: David W. Aha (aha '@' ics.uci.edu) (714) 856-8779")

print()

#target are patients with heart disease
target = df[(df['target'] == 1)]
#no_t are patients without heart disease
no_t = df[(df['target'] == 0)]

men = df[(df['sex'] == 1)]
women = df[(df['sex'] == 0)]
men_target = df[(df['sex'] == 1) & (df['target'] == 1)]
women_target = df[(df['sex'] == 0) & (df['target'] == 1)]
men_no_t = df[(df['sex'] == 1) & (df['target'] == 0)]
women_no_t = df[(df['sex'] == 0) & (df['target'] == 0)]

print("Patients: " + str(len(df['age'])))
print("Patients with heart disease: " + str(len(target)))
print("Patients without heart disease: " + str(len(no_t)))
print('Male patients: '+ str(len(men)))
print('Female patients: '+ str(len(women)))

print()

```

```python
print("Average age: " + str(np.mean(df['age'])))
print("Youngest patient: " + str(np.min(df['age'])))
print("Oldest patient: " + str(np.max(df['age'])))
print("Average age with heart disease: " + str(np.mean(target['age'])))
print("Average age with no heart disease: " + str(np.mean(no_t['age'])))
print()

print('Below are two subplots that looks at age has a correlation with heart disease.')

#subplot 1 - Patient Age
    #with heart disease
plt.subplot(2, 1, 1)
plt.plot(target['age'], '.', color='orange')
plt.title('With Heart Disease vs Without Heart Disease')
plt.xlabel('Patient Number')
plt.ylabel('Heart Disease: Age')

plt.subplot(2, 1, 2)
plt.plot(no_t['age'], '.', color='lightskyblue')
plt.xlabel('Patient Number')
plt.ylabel('No Heart Disease: Age')
plt.show()
    #without heart disease
print("It appears that age doesn't have a strong correlation with heart disease.")
print()
```

```python
print()

print('Men with heart disease: '+ str(len(men_target)) + '')
print('Women with heart disease: '+ str(len(women_target)))
print('Men without heart disease: '+ str(len(men_no_t)))
print('Women without heart disease: '+ str(len(women_no_t)))

print()

print("Below are two piecharts to have a more accurate idea how sex correlates with hear disease")
# Piechart Men with heart disease vs men without heart disease      
labels = 'Men With Heart Disease', 'Men Without Heart Disease'
sizes = [93, 72] #114
colors = ['orange','lightskyblue']
explode = (0.1, 0) 
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=90)
plt.axis('equal')
plt.show()

# Piechart women with heart disease vs women without heart disease  
labels = 'Women With Heart Disease', 'Women Without Heart Disease'
sizes = [72, 24]
colors = ['orange','lightskyblue']
explode = (0.1, 0) 
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=90)
plt.axis('equal')
plt.show()

print("What's interesting that is shown here is that a large majority 75% of women have heart disease while 45% of men have heart      disease. I'm not sure what to make of this. Maybe if we had a large data sample with more women I would be able to come to a   conclusion about this.")
print()



```

```python
plt.subplot(2, 1, 1)
plt.plot(target['chol'], '.', color='orange')
plt.title('Cholestrol Levels with Heart Disease vs Without Heart Disease')
plt.xlabel('Patient Number')
plt.ylabel('With Heart Disease')

plt.subplot(2, 1, 2)
plt.plot(no_t['chol'], '.', color='lightskyblue')
plt.xlabel('Patient Number')
plt.ylabel('Without Heart Disease')
plt.show()

print("Average cholestrol with heart disease: " + str(np.mean(target['chol'])))
print("Average cholestrol without heart disease: " + str(np.mean(no_t['chol'])))
```

```python
cp_hd = target[(target['cp'] >= 1)]
no_cp_hd = target[(target['cp'] == 0)]
cp_no_hd = no_t[(no_t['cp'] >= 1)]
no_cp_no_hd = no_t[(no_t['cp'] == 0)]
print('Patients with chest pain and heart disease: '+ str(len(cp_hd)))
print('Patients with no chest pain and heart disease: '+ str(len(no_cp_hd)))
print('Patients with chest pain and no heart disease: '+ str(len(cp_no_hd)))
print('Patients with no chest pain and no heart disease: '+ str(len(no_cp_no_hd)))

# Piechart Chest Pain: With and without Heart Disease     
labels = 'Chest Pain & Heart disease', 'Chest Pain & No Heart Disease'
sizes = [126, 34]
colors = ['orange', 'lightskyblue']
explode = (0.1, 0.0) 
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=90)
plt.axis('equal')
plt.show()

print("Chest pain seems to have a strong correlation with heart disease")
```
